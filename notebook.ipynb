## 1. Introduction
The World Happiness Report is a global survey that measures and ranks countries based on the happiness levels of their citizens. The rankings are calculated using factors such as GDP per capita, social support, healthy life expectancy, freedom to make life choices, generosity, and perceptions of corruption.

The report provides valuable insights into the well-being of people across different countries and helps researchers, governments, and organizations understand factors that contribute to happiness and quality of life.
Studying this dataset can help identify patterns and relationships between various social and economic factors and overall happiness levels.

## Why I Selected This Dataset

I selected the World Happiness Report dataset because I am interested in understanding what factors contribute to happiness in different countries. Happiness is influenced by various economic, social, and health-related factors, making it an interesting topic for data analysis.

This dataset provides an opportunity to explore how variables such as GDP, social support, life expectancy, freedom, generosity, and corruption relate to happiness scores. Understanding these relationships can provide useful insights into global well-being.

## Research Question 3
Does GDP per capita influence happiness?

```python
data[['GDP per capita','Score']].describe()
```

```python
data['GDP per capita'].corr(data['Score'])
```

# observation
The correlation coefficient between GDP per capita and Happiness Score is approximately 0.79, indicating a strong positive relationship between the two variables. Countries with higher GDP per capita generally tend to have higher happiness scores, while countries with lower GDP per capita tend to report lower happiness levels.

# interpretation
The strong positive correlation suggests that economic prosperity may play an important role in determining national happiness. A higher GDP per capita often reflects better living standards, access to resources, employment opportunities, and overall economic stability. However, since the correlation is not perfect, GDP alone cannot fully explain happiness, and other factors such as social support, health, and personal freedom may also contribute significantly.

# summary
The analysis revealed a strong positive relationship between GDP per capita and happiness score. Countries with stronger economies generally report higher levels of happiness. While GDP appears to be an important factor, additional variables should be examined to gain a more complete understanding of what influences happiness across countries.

# Research question 6
 Which factor has the strongest relationship with happiness?

```python
data.corr(numeric_only=True)['Score'].sort_values(ascending=False)
```

# observation
The correlation analysis shows that GDP per capita has the strongest positive relationship with happiness score (0.793883), followed closely by healthy life expectancy (0.779883) and social support (0.777058). Freedom to make life choices has a moderate positive relationship (0.566742), while perceptions of corruption shows a weaker positive relationship (0.385613). Generosity has the weakest relationship with happiness score (0.075824). The overall rank has a strong negative correlation because countries with lower rank numbers are ranked as happier countries.

# interpretation
The results suggest that economic prosperity, good health, and strong social support are the factors most strongly associated with happiness. Countries that provide better living standards, healthcare, and supportive social environments tend to report higher happiness levels. Freedom also contributes positively to happiness, although its impact appears less significant. The very weak correlation of generosity indicates that generosity alone may not be a major determinant of national happiness in this dataset.

# summary
Among all the factors analyzed, GDP per capita shows the strongest positive relationship with happiness score, followed closely by healthy life expectancy and social support. These findings indicate that economic stability, health, and social well-being are the most influential factors associated with happiness. Other factors such as freedom, perceptions of corruption, and generosity contribute to happiness but with comparatively weaker relationships.

```python
top10 = data.nlargest(10, 'Score')

top10.plot(
    x='Country or region',
    y='Score',
    kind='bar',
    figsize=(10,5)
)
```

The bar chart shows the ten countries with the highest happiness scores in the dataset. Finland achieved the highest happiness score, followed by Denmark, Norway, Iceland, and the Netherlands. The scores among the top-ranked countries are relatively close, indicating consistently high levels of happiness among these nations.

## interpretation
The concentration of several Northern European countries among the top positions suggests that factors such as economic stability, social support, quality healthcare, and personal freedom may contribute to higher happiness levels. However, further analysis is required to determine which factors have the strongest relationship with happiness.

## interpretation
Lower happiness scores may be associated with economic difficulties, weaker social support systems, health challenges, political instability, or limited opportunities. Additional analysis of the explanatory variables is necessary to identify the factors most strongly associated with lower happiness levels.

## visualization summary
The visualizations clearly demonstrate that happiness levels vary significantly across countries. While some countries consistently achieve high happiness scores, others face challenges that result in much lower levels of reported well-being. This variation motivates further investigation into the factors that influence happiness.

```python
correlations = data.corr(numeric_only=True)['Score'].drop(['Score','Overall rank'])

correlations.sort_values(ascending=False).plot(kind='bar')

plt.title('Factors Influencing Happiness Score')
plt.xlabel('Factors')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.show()
```

# observation
The bar chart shows that GDP per capita has the strongest positive correlation with happiness score (0.793883), followed closely by healthy life expectancy (0.779883) and social support (0.777058). Freedom to make life choices has a moderate positive correlation (0.566742), while perceptions of corruption has a weaker relationship (0.385613). Generosity shows the weakest correlation with happiness score (0.075824).

# interpretation
The visualization suggests that economic prosperity, health, and social support are the factors most strongly associated with happiness. Countries that provide better economic opportunities, healthcare, and supportive social environments tend to report higher happiness levels. While freedom also contributes positively to happiness, its influence appears less significant. Generosity alone does not appear to have a substantial impact on national happiness in this dataset.

# summary
The correlation comparison chart reveals that GDP per capita, healthy life expectancy, and social support are the strongest factors associated with happiness. Freedom to make life choices has a moderate influence, whereas perceptions of corruption and generosity show weaker relationships. Overall, the findings indicate that economic stability, good health, and strong social support systems are the key contributors to higher happiness levels across countries.
